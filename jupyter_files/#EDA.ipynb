```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

###################################################
intent_counts = data_aug.groupby(['partition', 'intent']).size().reset_index(name='count')

plt.figure(figsize=(10, 40))
plot = sns.barplot(x='count', y='emotion', hue='partition', data=intent_counts, orient='h')
plot.set(xlabel='Count', ylabel='Intent', title='Distribution of Intents in Each Partition')
plt.legend(title='Partition', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```
